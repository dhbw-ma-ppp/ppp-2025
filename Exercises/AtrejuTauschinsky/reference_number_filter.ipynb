```python
import string
from collections import Counter
import time
```

### short and readable but slow

```python
def number_is_valid(number):
    number_string = str(number)
    has_two = any(number_string.count(digit) == 2 for digit in number_string)
    is_monotonic = all(number_string[i] <= number_string[i + 1] for i in range(len(number_string) - 1))
    return is_monotonic and has_two
```

```python
t0 = time.monotonic()
result = sum(number_is_valid(x) for x in range(13456471, 58515929))
t1 = time.monotonic()

print(f'{result} calculated in {t1 - t0:.3}s')
```

### faster and somewhat simple still

```python
def generate_monotonic_digits(min, max):
    current = [x for x in str(min)]
    max = [x for x in str(max)]
    N = len(current)

    if N != len(max):  # could zero-pad min, but not worth it now
        raise NotImplementedError('min and max must have the same number of digits')
    
    # make the first number monotonic
    for i in range(N - 1):
        if current[i] > current[i + 1]:
            current[i+1:] = [current[i]]*(N - i - 1)
            break
    yield current[:]

    i = N - 1
    while i > 0:
        if current[i] < '9':
            fill_value = str(int(current[i]) + 1)
            current[i:] = [fill_value]*(N - i)
        else:
            while current[i] == '9':
                i -= 1
            fill_value = str(int(current[i]) + 1)
            current[i:] = [fill_value]*(N - i)
            i = N - 1

        is_smaller = False
        for j in range(N):
            if current[j] < max[j]:
                is_smaller = True
                break
            elif current[j] > max[j]:
                break
        if not is_smaller:
            break

        yield current[:]
            

def has_double_digit_count(x):
     for i in range(len(x) - 1):
        if x.count(x[i]) == 2:
            return True

def count_valid(min, max):
    valid_count = 0
    for x in generate_monotonic_digits(min, max):
        if has_double_digit_count(x):
            valid_count += 1
    return valid_count
```

```python
count_valid(13456471, 58515929)
```

```python
%%timeit
count_valid(13456471, 58515929)
```
