```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_csv = pd.read_csv("titanic.csv")
```

```python
sorted_by_age = df_csv.sort_values(by='Age', ascending=False)
display(sorted_by_age)
```

```python
nan_mask = sorted_by_age["Age"].isna()
display(nan_mask)
```

```python
sorted_by_age_no_nan = sorted_by_age.dropna()
display(sorted_by_age_no_nan)
```

```python
only_age = sorted_by_age_no_nan["Age"]
x_name = sorted_by_age_no_nan["Name"]
new_index = []
for i in range(183):
    new_index.append(i)
    


plt.figure(figsize=(8, 5))
plt.bar(new_index, only_age)

plt.xlabel("Personen")
plt.ylabel("Alter")
plt.title("Alter aller Personen, die ihr Alter angegeben haben: ")

average_age = np.mean(only_age)
plt.hlines(y = average_age, xmin = 0, xmax = new_index[-1], color = "red", label = "Durchschnittsalter")
plt.legend()

plt.show()
```

```python
survived_or_not = df_csv[["Survived", "Name"]]
```

```python
survived = survived_or_not[survived_or_not["Survived"] == 1]
not_survived = survived_or_not[survived_or_not["Survived"] == 0]
x1 = "Survived"
x2 = "Dead"

plt.figure(figsize=(8, 5))
plt.bar(x1, len(survived), color = "green")
plt.bar(x2, len(not_survived), color = "red")



plt.title("Anzahl der Toten und der Überlebenden: ")


plt.show()
```

```python

age_and_survived = df_csv[["Survived", "Age"]]
under_30 = age_and_survived[age_and_survived["Age"] <= 30]
survived_under_30 = under_30[under_30["Survived"] == 1]  
dead_under_30 = under_30[under_30["Survived"] == 0]      



under_50 = age_and_survived[(age_and_survived["Age"] > 30) & (age_and_survived["Age"] <= 50)]
survived_under_50 = under_50[under_50["Survived"] == 1]  
dead_under_50 = under_50[under_50["Survived"] == 0]      


over_50 = age_and_survived[(age_and_survived["Age"] > 50)]
survived_over_50 = over_50[over_50["Survived"] == 1]     
dead_over_50 = over_50[over_50["Survived"] == 0]         



x01 = "Survived"
x02 = "Dead"


fig01 = plt.figure(figsize=(10, 8))  
ax01 = plt.axes(projection='3d')


z1, z2, z3, z4, z5, z6 = len(survived_under_30), len(dead_under_30), len(survived_under_50), len(dead_under_50), len(survived_over_50), len(dead_over_50)
y01, y02, y03 = 1, 2, 3  


x_pos = [0, 1]  

ax01.bar3d(0, y01, 0, 0.5, 0.5, z5, color='green', alpha=0.8)  # Survived
ax01.bar3d(1, y01, 0, 0.5, 0.5, z6, color='red', alpha=0.8)    # Dead


ax01.bar3d(0, y02, 0, 0.5, 0.5, z3, color='green', alpha=0.8)  # Survived
ax01.bar3d(1, y02, 0, 0.5, 0.5, z4, color='red', alpha=0.8)    # Dead


ax01.bar3d(0, y03, 0, 0.5, 0.5, z1, color='green', alpha=0.8)  # Survived
ax01.bar3d(1, y03, 0, 0.5, 0.5, z2, color='red', alpha=0.8)    # Dead


ax01.set_title('Survival by Age Group', fontsize=14)
ax01.set_xlabel('Survival Status')
ax01.set_ylabel('Age Group')
ax01.set_zlabel('Number of Passengers')


ax01.set_xticks([0.25, 1.25])
ax01.set_xticklabels(['Survived', 'Dead'])
ax01.set_yticks([1, 2, 3])
ax01.set_yticklabels(['≤30', '31-50', '>50'])


from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='s', color='w', markerfacecolor='green', markersize=10, label='Survived'),
    Line2D([0], [0], marker='s', color='w', markerfacecolor='red', markersize=10, label='Dead')
]
ax01.legend(handles=legend_elements, loc='upper right')


plt.show()
```
