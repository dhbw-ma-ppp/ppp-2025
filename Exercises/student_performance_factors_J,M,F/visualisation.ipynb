```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('StudentPerformanceFactors.csv')
```

```python
print(data.shape)
print(data.isnull().sum())
```

```python
#score range
max_score = data['Exam_Score'].max()
min_score = data['Exam_Score'].min()
score_range = max_score - min_score
print(f"\nRange of Exam Scores: {score_range}")

# Average score
average_score = data['Exam_Score'].mean()
print(f"Average Exam Score: {average_score:.2f}")
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Exam_Score'], kde=True, color='green')
# Mittelwert einzeichnen
plt.axvline(data['Exam_Score'].mean(), color='red', linestyle='--', label=f"Mittelwert: {data['Exam_Score'].mean():.1f}")
plt.title('Verteilung der Prüfungsergebnisse')
plt.xlabel('Punkte')
plt.ylabel('Häufigkeit')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# regplot zeichnet Punkte und eine Trendlinie
sns.regplot(x='Hours_Studied', y='Exam_Score', data=data, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('Einfluss der Lernzeit auf die Note')
plt.xlabel('Lernstunden pro Woche')
plt.ylabel('Note')
plt.show()
```

```python
#scores by gender
plt.figure(figsize=(8, 6))  
sns.boxplot(x='Gender', y='Exam_Score', data=data)
plt.title('Exam Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Exam Score')
plt.show()
```

```python
#exam scores by sleep hours
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sleep_Hours', y='Exam_Score', data=data)
plt.title('Exam Scores by Sleep Hours')
plt.xlabel('Sleep Hours per Night')
plt.ylabel('Exam Score')
plt.show()
```

```python
# Reihenfolge definieren (musst du anpassen, je nachdem welche Werte in der Spalte stehen)
# Beispiel: Angenommen die Werte sind 'High School', 'College', 'Postgraduate'
edu_order = ['High School', 'College', 'Postgraduate'] 

plt.figure(figsize=(12, 6))
# Hier nutzen wir tatsächlich 'Parental_Education_Level'
sns.boxplot(x='Parental_Education_Level', y='Exam_Score', data=data, order=edu_order, palette="Set2")
plt.title('Noten nach Bildungsgrad der Eltern')
plt.xlabel('Bildungsgrad')
plt.ylabel('Note')
plt.xticks(rotation=45)
plt.show()
```
