```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('StudentPerformanceFactors.csv')
```

```python
print(data.shape)
print(data.isnull().sum())
```

```python
#score range
max_score = data['Exam_Score'].max()
min_score = data['Exam_Score'].min()
score_range = max_score - min_score
print(f"\nRange of Exam Scores: {score_range}")

# Average score
average_score = data['Exam_Score'].mean()
print(f"Average Exam Score: {average_score:.2f}")
```

```python
# Detailed distribution of exam scores
plt.figure(figsize=(8, 6))
sns.histplot(data['Exam_Score'], kde=True, color='green')
plt.title('Distribution of Exam Scores')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# scores by study time
plt.figure(figsize=(10, 6))
sns.boxplot(x='Hours_Studied', y='Exam_Score', data=data)
plt.title('Exam Scores by Study Time')
plt.xlabel('Study Time (hours per week)')
plt.ylabel('Exam Score')
plt.show()
```

```python
#scores by gender
plt.figure(figsize=(8, 6))  
sns.boxplot(x='Gender', y='Exam_Score', data=data)
plt.title('Exam Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Exam Score')
plt.show()
```

```python
#exam scores by sleep hours
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sleep_Hours', y='Exam_Score', data=data)
plt.title('Exam Scores by Sleep Hours')
plt.xlabel('Sleep Hours per Night')
plt.ylabel('Exam Score')
plt.show()
```

```python
#scores by parental involvement
plt.figure(figsize=(12, 6))
sns.boxplot(x='Parental_Involvement', y='Exam_Score', data=data)
plt.title('Exam Scores by Parental Involvement')
plt.xlabel('Parental Involvement')
plt.ylabel('Exam Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Select only numerical columns
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Heatmap showing correlations between numerical features in the dataset
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
