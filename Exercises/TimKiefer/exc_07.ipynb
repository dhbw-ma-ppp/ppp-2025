```python
with open('data/exercise_cave.txt') as f:
    grid = [[int(x) for x in line.strip()] for line in f]
```

```python
# --- 1. Version ---
i = 0
j = 0
counter = 0

while j != 99 and i != 99:
    x = grid[i][j+1]
    y = grid[i+1][j]

    if x > y:
        i += 1
    else:  
        j += 1

    a = int(grid[i][j])
    counter += a

    # Prints to understand whats not working...
    #print(grid[i][j])
    #print(i,j)
    #print(f"Current Count: {counter}")

print(f"Final Count: {counter}")
```

```python
# Needed Corrections // Learning by doing: 
# Wrong Endpoint! Its a need to end at the right bottom!
# If the Side of the Grid is reached it fails! 
# More than just 2 Directions!
# The smallest way now isn't the lowest risk overall... a way of valueing the longterm move maybe per heap?
# A way of avoiding to visit smth more than once... the infinity loop wasn't nice.

# --- 2. Version ----
import heapq

#new formating...
rows = len(grid)
cols = len(grid[0])

# Priority Queue
pq = [(0, 0, 0)] # (Costs by now, x, y)

# Memory of Visits
visited = set()


directions = [
    (0, 1),  # right
    (0, -1), # left
    (1, 0),  # down 
    (-1, 0)  # up
]

print("Finding the lowst Risk...")

while pq:
    # taking way with lowest cost point
    cost, x, y = heapq.heappop(pq)

    # are we there yet?
    if x == rows - 1 and y == cols - 1:
        print(f"Final Count (Lowest Risk): {cost}")
        break

    # visited?
    if (x, y) in visited:
        continue
    visited.add((x, y))

    # checking out the others
    for dx, dy in directions:
        nx, ny = x + dx, y + dy

        # Is it still in the grid
        if 0 <= nx < rows and 0 <= ny < cols:
            new_cost = cost + grid[nx][ny]
            
            # securing newst step
            heapq.heappush(pq, (new_cost, nx, ny))
```
